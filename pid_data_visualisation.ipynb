```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from CSV
df = pd.read_csv('floppydata.csv', header=None)

# Ensure only rows with exactly 6 columns are kept
df = df[df.apply(lambda x: len(x.dropna()) == 6, axis=1)]

# Assign correct column names
df.columns = ['integral', 'derivative', 'tilterror','Time (s)' ,'pidOutput', 'pitch', ]

# Convert columns to numeric
for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')


# Create figure with 5 subplots
fig, axs = plt.subplots(5, 1, figsize=(10, 15), sharex=True)

# Plot each variable separately
plot_data = {
    'PID Output': 'pidOutput',
    'Tilt Angle': 'pitch',
    'Integral': 'integral',
    'Derivative': 'derivative',
    'Tilt Error': 'tilterror'
}
colors = ['b', 'r', 'g', 'm', 'c']

for ax, (title, col), color in zip(axs, plot_data.items(), colors):
    ax.plot(df['Time (s)'], df[col], label=title, color=color)
    ax.set_ylabel(title)
    ax.set_title(f'{title} vs Time')
    ax.legend()
    ax.grid()

# Set common X-axis label
axs[-1].set_xlabel('Time (s)')

# Adjust layout for better spacing
plt.tight_layout()

# Show the plots
plt.show()
```
